### 到底删除几个文本向量好呢？
思路：
- 剩余的向量个数应大于等于总个数的50%
- 删除之后效果应该更好，将删除方式记录、将计算出来的最好的中心记录

### 直接先尝试看一下各个文本中心的情况吧！


```python
# 先定义相似度（夹角余弦）的计算公式：
def cos_sim(a,b):
    cos = np.dot(a,b)/(np.linalg.norm(a)*(np.linalg.norm(b)))
    return 0.5 + 0.5 * cos
```

```python
t0 = time.time()
keys = list(goal_centers.keys())
goals_sim = pd.DataFrame(index=keys,columns=keys)
for r in range(len(keys)):
    for c in range(r):
        goals_sim.iloc[r,c] = cos_sim(goal_centers[keys[r]],goal_centers[keys[c]])
print('共耗时',(time.time()-t0)/60,'min')
```
