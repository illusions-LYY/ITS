### jieba分词思路：
- 先把所有二级目标分词，然后取set，查看有没有分错的词，组成一个词表，load进jieba用户词典中
- 再用load进了用户自定义词典的词表进行二次jieba分词，得到分词结果
- 根据二次分词结果建立停用词表
- 遍历并删除停用词，完成jieba分词

```python
def is_number(s):
    try:
        float(s)
        return False
    except ValueError:
        pass
 
    try:
        import unicodedata
        unicodedata.numeric(s)
        return False
    except (TypeError, ValueError):
        pass
 
    return True
```

```python
def seg_words(ss,stopwords,user_latex):
    # 载入用户自定义词典
    jieba.load_userdict("/Users/shen-pc/Desktop/WORK/ITS/My idea/my jieba list.txt")
    # 正则匹配字符串中的代数式or字母，大小写均匹配，然后删除这些字母中的LaTeX关键词，用作后面筛选
    pattern = re.compile(r'[\d]*[a-zA-Z]+')
    letter = pattern.findall(ss)
    letter_shelter = set(letter) - set(user_latex)
    # jieba分词
    ss_list = jieba.cut(ss.strip())
    outstr = ''
    for word in ss_list:
        # 去除每一个分出词的左右空格，防止漏网之鱼出现
        word_s = word.strip()
        # 这么多if：word在停用词表里、在正则找到的代数式or字母里、是空字符串、是制表符、是数字的，均删除
        if word_s not in stopwords:
            if word_s not in letter_shelter:
                if word_s != '' and word_s != '\t' and is_number(word_s):  
                    outstr += word_s  
                    outstr += " "
    return outstr  
```
